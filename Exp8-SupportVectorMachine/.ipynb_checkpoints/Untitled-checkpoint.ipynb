```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.svm import SVC
```

```python
df = pd.read_csv('data.csv')
df['rating_difference'] = df['white_rating'] - df['black_rating']
df['white_win'] = df['winner'].apply(lambda x : 1 if x=='white' else 0)
```

```python
df.iloc[:,[0,1,5,6,8,9,10,11,13,16,17]]
```

```python
X = df[['rating_difference', 'turns']]
Y = df['white_win'].values
# X_train, X_test, Y_train, Y_test, clf = fitting(X, Y, 1, 'scale')
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
model = SVC(kernel = 'rbf')
clf = model.fit(X_train, Y_train)
```

```python
pred_labels_train = model.predict(X_train)
pred_labels_test = model.predict(X_test)
```

```python
# Evaluation on Test Data
score_test = model.score(X_test, Y_test)
print("Accuracy Score: ", score_test)
print(classification_report(Y_test, pred_labels_test))
```

```python
# Evaluation on Training Data
score_train = model.score(X_train, Y_train)
print("Accuracy Score: ", score_train)
print(classification_report(Y_train, pred_labels_train))
```
